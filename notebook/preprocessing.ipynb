```python
from app.rag.loaders.document_loader import DocumentLoader
from app.rag.preprocessing.preprocessor import Preprocessor
from app.rag.chunking.chunk_manager import ChunkManager
from app.rag.embeddings.embedding_manager import EmbeddingManager

loader = DocumentLoader("../documents")

documents = loader.load()

preprocessor = Preprocessor()

processed_documents = preprocessor.process(documents)

chunk_manager = ChunkManager()

chunked_documents = chunk_manager.process(processed_documents)

embedding_manager = EmbeddingManager()

embedded_documents, embeddings = embedding_manager.process(
    chunked_documents
)
```

```python
len(embedded_documents)
len(embeddings[0])
embedded_documents[0].metadata

query_embedding = embedding_manager.embed_query(
    "How authentication works?"
)

len(query_embedding)
```

```python
from app.rag.loaders.document_loader import DocumentLoader
from app.rag.preprocessing.preprocessor import Preprocessor
from app.rag.chunking.chunk_manager import ChunkManager
from app.rag.vectordb.vector_store_manager import VectorStoreManager

loader = DocumentLoader("../documents")
documents = loader.load()

preprocessor = Preprocessor()
processed_documents = preprocessor.process(documents)

chunk_manager = ChunkManager()
chunked_documents = chunk_manager.process(processed_documents)

vector_manager = VectorStoreManager()

vector_manager.index_documents(chunked_documents)
```
