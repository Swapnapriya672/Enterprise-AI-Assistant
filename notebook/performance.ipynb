```python
import time
from statistics import mean

from app.agent.agent_manager import AgentManager

agent = AgentManager()

TEST_QUERIES = [

    ("SQL", "Show all applications"),

    ("RAG", "Explain authentication"),

    ("HYBRID", "Show all applications and explain authentication"),

    ("IRRELEVANT", "What is cricket?")

]

ITERATIONS = 5

print("=" * 100)
print("ENTERPRISE AI ASSISTANT PERFORMANCE TEST")
print("=" * 100)

results = []

for query_type, query in TEST_QUERIES:

    print(f"\n{query_type}")
    print("-" * 100)
    print(query)

    timings = []

    for i in range(ITERATIONS):

        start = time.perf_counter()

        response = agent.run(query)

        end = time.perf_counter()

        elapsed = end - start

        timings.append(elapsed)

        print(
            f"Run {i+1}: {elapsed:.3f} sec"
        )

    avg = mean(timings)

    minimum = min(timings)

    maximum = max(timings)

    results.append({

        "type": query_type,

        "query": query,

        "avg": avg,

        "min": minimum,

        "max": maximum

    })

print("\n")
print("=" * 100)
print("SUMMARY")
print("=" * 100)

for result in results:

    print(f"\n{result['type']}")

    print(f"Query : {result['query']}")

    print(f"Average : {result['avg']:.3f} sec")

    print(f"Minimum : {result['min']:.3f} sec")

    print(f"Maximum : {result['max']:.3f} sec")
```
