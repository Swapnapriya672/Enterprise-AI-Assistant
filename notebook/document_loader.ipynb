```python
from pathlib import Path
import json
import pandas as pd

from langchain_community.document_loaders import (
    PyPDFLoader,
    Docx2txtLoader,
    TextLoader,
)

from langchain_core.documents import Document

DOCUMENTS_PATH = Path("../documents")

DOCUMENTS_PATH
```

```python
files = list(DOCUMENTS_PATH.iterdir())

print(f"Total Files : {len(files)}")

for file in files:
    print(file.name)
```

```python
pdf_documents = []

pdf_files = DOCUMENTS_PATH.glob("*.pdf")

for pdf_file in pdf_files:

    loader = PyPDFLoader(str(pdf_file))

    docs = loader.load()

    for doc in docs:

        doc.metadata.update(
            {
                "file_name": pdf_file.name,
                "file_type": "pdf",
                "file_extension": pdf_file.suffix,
                "file_path": str(pdf_file),
                "document_type": "Technical Document"
            }
        )

    pdf_documents.extend(docs)

print(f"Loaded {len(pdf_documents)} PDF pages.")
```

```python
docx_documents = []

docx_files = DOCUMENTS_PATH.glob("*.docx")

for docx_file in docx_files:

    loader = Docx2txtLoader(str(docx_file))

    docs = loader.load()

    for doc in docs:

        doc.metadata.update(
            {
                "file_name": docx_file.name,
                "file_type": "docx",
                "file_extension": docx_file.suffix,
                "file_path": str(docx_file),
                "document_type": "Word Document"
            }
        )

    docx_documents.extend(docs)

print(f"Loaded {len(docx_documents)} Word documents.")
print(docx_documents[0].metadata)
```

```python
text_documents = []

txt_files = DOCUMENTS_PATH.glob("*.txt")

for txt_file in txt_files:

    loader = TextLoader(str(txt_file))

    docs = loader.load()

    for doc in docs:

        doc.metadata.update(
            {
                "file_name": txt_file.name,
                "file_type": "txt",
                "file_extension": txt_file.suffix,
                "file_path": str(txt_file),
                "document_type": "Text Document"
            }
        )

    text_documents.extend(docs)

print(f"Loaded {len(text_documents)} Text documents.")
```

```python
markdown_documents = []

markdown_files = DOCUMENTS_PATH.glob("*.md")

for markdown_file in markdown_files:

    with open(markdown_file, encoding="utf-8") as file:

        content = file.read()

        markdown_documents.append(

            Document(

                page_content=content,

                metadata={
                    "file_name": markdown_file.name,
                    "file_type": "markdown",
                    "file_extension": markdown_file.suffix,
                    "file_path": str(markdown_file),
                    "document_type": "Markdown Document"
                }

            )

        )

print(f"Loaded {len(markdown_documents)} Markdown documents.")
```

```python
csv_documents = []

csv_files = DOCUMENTS_PATH.glob("*.csv")

for csv_file in csv_files:

    dataframe = pd.read_csv(csv_file)

    document = Document(

        page_content=dataframe.to_csv(index=False),

        metadata={
            "file_name": csv_file.name,
            "file_type": "csv",
            "file_extension": csv_file.suffix,
            "file_path": str(csv_file),
            "document_type": "CSV Document",
            "rows": len(dataframe),
            "columns": list(dataframe.columns)
        }

    )

    csv_documents.append(document)

print(f"Loaded {len(csv_documents)} CSV document(s).")
```

```python
excel_documents = []

excel_files = DOCUMENTS_PATH.glob("*.xlsx")

for excel_file in excel_files:

    excel_data = pd.read_excel(
        excel_file,
        sheet_name=None
    )

    for sheet_name, dataframe in excel_data.items():

        document = Document(

            page_content=dataframe.to_csv(index=False),

            metadata={
                "file_name": excel_file.name,
                "file_type": "excel",
                "file_extension": excel_file.suffix,
                "file_path": str(excel_file),
                "sheet_name": sheet_name,
                "document_type": "Excel Document",
                "rows": len(dataframe),
                "columns": list(dataframe.columns)
            }

        )

        excel_documents.append(document)

print(f"Loaded {len(excel_documents)} Excel sheet(s).")
```

```python
json_documents = []

json_files = DOCUMENTS_PATH.glob("*.json")

for json_file in json_files:

    with open(json_file, "r", encoding="utf-8") as file:

        json_data = json.load(file)

    document = Document(

        page_content=json.dumps(
            json_data,
            indent=4
        ),

        metadata={
            "file_name": json_file.name,
            "file_type": "json",
            "file_extension": json_file.suffix,
            "file_path": str(json_file),
            "document_type": "JSON Document",
            "total_keys": len(json_data) if isinstance(json_data, dict) else None
        }

    )

    json_documents.append(document)

print(f"Loaded {len(json_documents)} JSON document(s).")
```

```python
documents = []

documents.extend(pdf_documents)
documents.extend(docx_documents)
documents.extend(text_documents)
documents.extend(markdown_documents)
documents.extend(csv_documents)
documents.extend(excel_documents)
documents.extend(json_documents)

print(f"Total Documents Loaded : {len(documents)}")
print(documents[0].metadata)
```

```python
for document in documents:

    print(document.metadata)

    print("-" * 100)
```

```python
file_types = {}

for document in documents:

    file_type = document.metadata.get("file_type", "Unknown")

    file_types[file_type] = file_types.get(file_type, 0) + 1

print(file_types)
```
